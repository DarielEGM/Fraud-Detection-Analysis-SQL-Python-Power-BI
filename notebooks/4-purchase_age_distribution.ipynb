```python
# Importing Libraries
from data_loader import load_data
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt  

df = load_data()
df_temp = df.copy()
df_temp['payment_method'].fillna('Unknown', inplace=True)
```

```python
# Create a figure with two subplots to visualize the distribution of customer_age and purchase_amount
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))

# Boxplot for customer age on the first subplot (ax[0])
sns.boxplot(data=df_temp, y='customer_age', ax=ax[0])
ax[0].set_title('Distribution of Customer Age')
ax[0].set_ylabel('Customer Age')
ax[0].set_ylim(15, 75)
ax[0].set_yticks(range(15, 75, 5))

# Boxplot for purchase amount on the second subplot (ax[1])
sns.boxplot(data=df_temp, y='purchase_amount', ax=ax[1])
ax[1].set_title('Distribution of Purchase Amount')
ax[1].set_ylabel('Purchase Amount')
ax[1].set_ylim(25, 325)
ax[1].set_yticks(range(25, 325, 25))

# Adjust plot layout to prevent elements from overlapping
plt.tight_layout()

# Display the plot
plt.show()
```
