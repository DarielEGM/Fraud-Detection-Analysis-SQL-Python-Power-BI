```python
# Importing Libraries
from data_loader import load_data
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt  

df = load_data()
df_temp = df.copy()
df_temp['payment_method'].fillna('Unknown', inplace=True)
```

```python
# Count the number of transactions per location
df_transaction_location = df_temp.groupby('location')['transaction_id'].count().reset_index(name='count_transactions')

# Sort the data to make the chart easier to interpret
df_transaction_location = df_transaction_location.sort_values(by='count_transactions', ascending=False)

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=df_transaction_location, x='location', y='count_transactions')

# Add titles and labels for the chart
plt.title('Distribution of Transactions by Location')
plt.xlabel('Location')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
